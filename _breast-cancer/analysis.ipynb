```python
smaller_data = df.loc[:, ['radius_mean', 'perimeter_mean', 'area_mean',
                          'smoothness_mean', 'symmetry_mean', 'fractal_dimension_mean']]
bins = 12
plt.figure(figsize=(10, 10))
for idx, feature in enumerate(smaller_data):
    rows = 5
    cols = 2
    plt.subplot(rows, cols, idx+1)

    sns.distplot(df[df['diagnosis']=='M'][feature], bins=bins, color='red', label='M')
    sns.distplot(df[df['diagnosis']=='B'][feature], bins=bins, color='blue', label='B')
    plt.legend(loc='upper right')

plt.show()
```

```python
smaller_data = df.loc[:, ['radius_mean', 'perimeter_mean', 'area_mean',
                          'smoothness_mean', 'symmetry_mean', 'fractal_dimension_mean']]
bins = 12
plt.figure(figsize=(10, 10))
for idx, feature in enumerate(smaller_data):
    rows = 6
    cols = 2
    plt.subplot(rows, cols, idx+1)

    sns.distplot(df[df['diagnosis'] == 'M'][feature],
                 bins=bins, color='red', label='M')
    sns.distplot(df[df['diagnosis'] == 'B'][feature],
                 bins=bins, color='blue', label='B')
    plt.legend(loc='upper right')

plt.show()
```
