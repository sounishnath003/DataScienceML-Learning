```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv("../metrics.csv")

# Plot training loss and accuracy
plt.figure(figsize=(10, 6))
plt.plot(data["epoch"], data["train_loss"], label="Train Loss", marker="o")
plt.plot(data["epoch"], data["train_acc"], label="Train Accuracy", marker="s")
plt.xlabel("Epoch")
plt.ylabel("Loss/Accuracy")
plt.title("Training Loss and Accuracy vs. Epoch")
plt.legend()
plt.grid(True)

# Plot validation loss and accuracy
plt.figure(figsize=(10, 6))
plt.plot(data["epoch"], data["loss"], label="Validation Loss", marker="o")
plt.plot(data["epoch"], data["val_acc"], label="Validation Accuracy", marker="s")
plt.xlabel("Epoch")
plt.ylabel("Loss/Accuracy")
plt.title("Validation Loss and Accuracy vs. Epoch")
plt.legend()
plt.grid(True)

# Show the plots
plt.show()
```
